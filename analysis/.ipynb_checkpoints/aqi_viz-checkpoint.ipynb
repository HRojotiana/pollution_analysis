```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from extraction.extract import extract_from_database
```

```python
df = extract_from_database('air_pollutions')
non_numerical_col = ['Date', 'Location']
numerical_col = [col for col in df.columns if col not in non_numerical_col]
numerical_df = df[numerical_col]
```

```python
#Aqi distribution
plt.figure(figsize=(10, 6))
sns.histplot(numerical_df['aqi'], kde=True)
plt.title('aqi Distribution')
plt.xlabel('aqi')
plt.ylabel('Fréquences')
plt.show()
```

```python
sns.boxplot(x=numerical_df['aqi'])
plt.xlim(0)
plt.title('aqi boxplot')
plt.xlabel('aqi')
plt.show()
```

```python
percentiles = np.percentile(numerical_df['aqi'], [0, 25, 50, 75, 100])

sns.boxplot(x=numerical_df['aqi'])
plt.xlim(0)
plt.title('aqi boxplot')
plt.xlabel('aqi')

for p in percentiles:
    plt.text(p, 0.02, f'{p:.2f}', horizontalalignment='center')

plt.show()
```

```python
# aqi per location
sns.barplot(x='Location', y='aqi', data=df, errorbar=None)
plt.title('aqi per location')
plt.xlabel('Location')
plt.ylabel('aqi')
plt.show()
```

```python
# aqi correlation
# Correlation matrix
correlation_matrix = numerical_df.corr()

#Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation heatmap')
plt.show()
```
